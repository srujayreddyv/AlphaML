##Indentation

```python
#build a calculator
#input: x, y, operations (*, +, -)
#output: number or invalid

x = input("first number:")
y = input("second number:")
op = input("pick operation: *, +, -")
```

```python
print(type(x))
```

```python
x = int(x)
y = int(y)
print(type(x), type(y))
```

```python
if op == '*':
  result = x*y;

elif op == '+':
  result = x+y;

elif op == '-':
  result = x-y;

else:
  result = "invalid";
```

```python
print("result:", result)
```
